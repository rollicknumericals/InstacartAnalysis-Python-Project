## 6. Conduct exploratory analysis of customer demographics to inform targeted marketing campaigns. Determine if there's a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents.

```python
# Create random seed
np.random.seed(9)
```

```python
# Create dev to store boolean values
dev = np.random.rand(len(df)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe - 'big'
big = df[dev]
```

```python
# Store 30% of the sample in the dataframe - 'small'
small = df[~dev]
```

```python
# Check for expected results
len(df)
```

```python
len(big) + len(small)
```

```python
# Reduce samples to only columns necessary for line chart
df_2 = small[['age', 'dependents']]
```

```python
# Create line chart
line = sns.lineplot(data = df_2, x = 'age', y = 'dependents')
```

##### No obvious pattern emerges in this line chart suggesting that there is not a strong connection between age and number of dependents.

## 7. Explore whether there's a connection between age and spending power (income) by creating a scatterplot.

```python
# Reduce samples to only columns necessary for scatterplot
df_3 = small [['age','income']]
```

```python
# Create scatterplot
scatter = sns.scatterplot(x = 'age', y = 'income', data = df_3)
```

##### The scatterplot shows there is a high-level concentration among all ages that make at least 20k/yr. There is a low-level concentration of those aged 20-40 that make between 20k/yr and 30k/yr. There is a medium to high-level concentration of those aged 40 and up that make between 20k/yr and 30k/yr. Those aged 20-40 have a medium to low-level concentration that make between 30k/yr and 40k/yr. Those aged 20-40 do not make above 40k/yr. There is a low-level concentration of those 40 and up that make between 30k/yr and 60k/yr. From these observations we can reliably say that income tends to rise after the age of 40.
