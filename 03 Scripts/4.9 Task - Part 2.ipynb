### 3. Create histogram of 'hour_ordered' column to provide IC senior stakeholders with descriptive findings about sales

```python
plt.title('Orders Placed by Hour of the Day')
hist = df['hour_ordered'].plot.hist(bins = 70)
plt.xlabel('Hour Ordered')
plt.ylabel('Frequency of Orders Placed x 1 million')
```

##### The vast majority of orders are placed between the hours of 10am and 5pm. Ordering activity is practically non-existent between the hours of 2am and 6am. There is a sharp rise between the hours of 7am and 9am. There is a gradual fall between the hours of 6pm and 1am. 10am is the most active ordering hour. 4am is the least active.

### 4. Create bar chart from 'loyalty_flag' column for marketing team

### 5. Check for difference in expenditure depending on the hour

```python
# Create random seed
np.random.seed(9)
```

```python
# Create dev to store boolean values
dev = np.random.rand(len(df)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the dataframe - 'big'
big = df[dev]
```

```python
# Store 30% of the sample in the dataframe - 'small'
small = df[~dev]
```

```python
# Check for expected results
len(df)
```

```python
len(big) + len(small)
```

```python
# Reduce samples to only columns necessary for line chart
df_2 = small[['hour_ordered', 'prices']]
```

```python
# Create line chart
plt.title('Order Expenditure by Hour of the Day') 
line = sns.lineplot(data = df_2, x = 'hour_ordered', y = 'prices')
plt.xlabel('Hour Ordered')
plt.ylabel('Expenditure Per Order')
```
