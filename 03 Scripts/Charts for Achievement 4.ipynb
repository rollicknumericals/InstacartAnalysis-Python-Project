### Loyalty charts

```python
# Construct loyalty_flag chart
plt.figure(figsize = (10,6))
plt.subplots_adjust(left = 0.3)
plt.title('Instacart Customer Loyalty Distribution')
loyalty_pie = df['loyalty_flag'].value_counts().plot.pie(autopct = ('%1.1f%%'), colors = ['salmon', 'lightgreen', 'plum'])
plt.ylabel('')
```

###### 'Loyal customers' have placed more than 40 orders with IC. 'Regular customers' have placed between 11 and 40 orders. 'New customers have placed between 5 and 10.

```python
# Export pie chart
loyalty_pie.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'loyalty_pie.png'))
```

```python
# Filter for 'Loyal customer' within 'loyalty_flag
loyal_custs = df[df['loyalty_flag'] == 'Loyal customer']
```

```python
# Count orders per department
dept_counts_loyal = loyal_custs['department'].value_counts()
```

```python
# Identify top 10 departments for 'Loyal customer'
top_depts_loyal = dept_counts_loyal.head(10)
```

```python
top_depts_loyal
```

```python
# Create bar chart of top 10 departments for loyal customers
plt.figure(figsize = (10,6))
plt.subplots_adjust(left = 0.3)
plt.title('Top 10 Departments for Loyal Customers')
top_depts_loyal_bar = top_depts_loyal.sort_values().plot.barh(color = 'orchid')
plt.xlabel('Number of Orders in each Department x 1 million')
plt.ylabel('Department Name')
```

```python
# Filter for 'Regular customer' within 'loyalty_flag
reg_custs = df[df['loyalty_flag'] == 'Regular customer']
```

```python
# Count orders per department
dept_counts_reg = reg_custs['department'].value_counts()
```

```python
# Identify top 10 departments for 'Regular customer'
top_depts_reg = dept_counts_reg.head(10)
```

```python
top_depts_reg
```

```python
# Create bar chart of top 10 departments for regular customers
plt.figure(figsize = (10,6))
plt.subplots_adjust(left = 0.3)
plt.title('Top 10 Departments for Regular Customers')
top_depts_reg_bar = top_depts_reg.sort_values().plot.barh(color = 'mediumvioletred')
plt.xlabel('Number of Orders in each Department x 1 million')
plt.ylabel('Department Name')
```

```python
# Count orders per department
dept_counts_new = new_custs['department'].value_counts()
```
